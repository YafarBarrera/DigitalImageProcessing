```python
image=cv2.imread("WIN_20240229_19_22_48_Pro.JPG", cv2.IMREAD_COLOR)
plt.figure()

plt.subplot(141) #Matriz de una fila, 3 items y que empieza en 1
plt.title('Original')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(142)
plt.title('Red Channel')
plt.imshow(image[:,:,0])
plt.axis('off')

plt.subplot(143)
plt.title('Green Channel')
plt.imshow(image[:,:,1],cmap='Reds')
plt.axis('off')

plt.subplot(144)
plt.title('Blue Channel')
plt.imshow(image[:,:,2],cmap='Blues')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread("WIN_20240229_19_22_48_Pro.JPG", cv2.IMREAD_GRAYSCALE)
original_img = image.dtype # Guardamos la profundidad original de nuestra imagen.
print("Profundidad original de color")
img_16bits = np.uint16(image) #Transformamos nuestra imagen a 16 Bits
new_img = img_16bits.dtype
print("Profundidad de color nueva")
cv2.imwrite('new_img.jpg',img_16bits)
img_saved = cv2.imread('new_img.jpg', cv2.IMREAD_UNCHANGED)
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.title('Image Original')
plt.imshow(image, cmap='gray')

plt.subplot(122)
plt.title('Image Changed')
plt.imshow(img_saved, cmap='gray')
```
