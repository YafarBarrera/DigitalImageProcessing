```python
import cv2
import matplotlib.pyplot as plt

def histograma(imagen):
    histo = cv2.calcHist([imagen],[0],None,[128],[0,128])

    plt.figure(figsize=(12, 4))
    plt.plot(histo, color = "purple")
    plt.title('HISTOGRAMA')
    plt.xlabel('Intensidad de pixel')
    plt.ylabel('Frecuencia')

img = cv2.imread('WIN_20240229_19_22_48_Pro.JPG', 0)
histograma(img)
```

Binarización

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('WIN_20240229_19_22_48_Pro.JPG', 0)
_,umbralbi = cv2.threshold(img,127,255,cv2.THRESH_BINARY) #Aplicando binarización
_,umbral_inverso = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV) #Aplicando binarización inversa

img2 = cv2.imread('Shar.JPG', 0)
_,umbralbi2 = cv2.threshold(img2,128,255,cv2.THRESH_BINARY) #Aplicando binarización
_,umbral_inverso2 = cv2.threshold(img2,128,255,cv2.THRESH_BINARY_INV) #Aplicando binarización inversa

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.imshow(img,cmap="gray")
plt.title("Original")

plt.subplot(132)
plt.imshow(umbralbi,cmap="gray")
plt.title("Binary")

plt.subplot(133)
plt.imshow(umbral_inverso,cmap="gray")
plt.title("Binary Inverse")

plt.show

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.imshow(img2,cmap="gray")
plt.title("Original")

plt.subplot(132)
plt.imshow(umbralbi2,cmap="gray")
plt.title("Binary")

plt.subplot(133)
plt.imshow(umbral_inverso2,cmap="gray")
plt.title("Binary Inverse")

plt.show
```

Umbrales Adaptativos

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("Shar.JPG",0)
umbral_adaptive = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,9,3)
umbral_adaptive_gaus = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,9,3)

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.imshow(img,cmap="gray")
plt.title("Original")

plt.subplot(132)
plt.imshow(umbral_adaptive,cmap="gray")
plt.title("Adaptive Umbral")

plt.subplot(133)
plt.imshow(umbral_adaptive_gaus,cmap="gray")
plt.title("Gaussian Umbral")

plt.show
```
