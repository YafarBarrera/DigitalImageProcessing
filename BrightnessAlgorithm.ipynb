Algoritmo de brillo.

```python
import cv2
import matplotlib.pyplot as plt
import copy

img = cv2.imread('Shar.jpg')

img_gray = cv2.imread('Shar.jpg', cv2.IMREAD_GRAYSCALE)

def brightness(img,factor):
    hight = len(img)
    width = len(img[0])
    new_img=copy.deepcopy(img)
    for i in range(0,hight):
        for j in range(0,width):
            if(img[i][j]+factor>255):
                new_img[i][j]=255
            elif(img[i][j]+factor<0):
                new_img[i][j]=0
            else:
                new_img[i][j]=new_img[i][j]+factor
    return new_img

def brightness_color(img,factor):
    hight = len(img)
    width = len(img[0])
    new_img=copy.deepcopy(img)
    for i in range(0,hight):
        for j in range(0,width):
            for k in range(0,3):
                if(img[i][j][k]+factor>255):
                    new_img[i][j][k]=255
                elif(img[i][j][k]+factor<0):
                    new_img[i][j][k]=0
                else:
                    new_img[i][j][k]=new_img[i][j][k]+factor
    return new_img

img_brightened_high=brightness(img_gray,69)
img_brightened_low=brightness(img_gray,-36)

img_brightened_color_high=brightness_color(img,69)
img_brightened_color_low=brightness_color(img,-36)

plt.figure(figsize=(12,3))

plt.subplot(131)
plt.imshow(img_gray,cmap="gray")
plt.title("Normal Brightness")

plt.subplot(132)
plt.imshow(img_brightened_high, cmap="gray")
plt.title("High Brightness")

plt.subplot(133)
plt.imshow(img_brightened_low, cmap="gray")
plt.title("Low Brightness")

plt.show()

plt.figure(figsize=(12,3))

plt.subplot(131)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Normal Brightness Color")

plt.subplot(132)
plt.imshow(cv2.cvtColor(img_brightened_color_high, cv2.COLOR_BGR2RGB))
plt.title("High Brightness Color")

plt.subplot(133)
plt.imshow(cv2.cvtColor(img_brightened_color_low, cv2.COLOR_BGR2RGB))
plt.title("Low Brightness Color")

plt.show()
```
