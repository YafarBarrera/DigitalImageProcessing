Ecualizaci√≥n de Histograma

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread('Shar2.jpg', cv2.IMREAD_GRAYSCALE)

histogram,_ = np.histogram(img_gray.flatten(),256,[0,256])
function_distrib=histogram.cumsum()
histo_normalized= function_distrib*histogram.max()/function_distrib.max()    #resultante de la funci√≥n integral
histo_equalized=cv2.equalizeHist(img_gray)

plt.figure(figsize=(12, 9))

plt.subplot(221)
plt.imshow(img_gray,cmap="gray")
plt.title("Imagen Original")

plt.subplot(222)
plt.plot(histogram, color = "purple")
plt.title('HISTOGRAMA')
plt.xlabel('Intensidad de pixel')
plt.ylabel('Frecuencia')

plt.subplot(223)
plt.plot(histo_normalized, color = "purple")
plt.title('HISTOGRAMA NORMALIZADO')
plt.xlabel('Intensidad de pixel')
plt.ylabel('Frecuencia')

plt.subplot(224)
plt.plot(histo_equalized, color = "purple")
plt.title('HISTOGRAMA EQUALIZADO')
plt.xlabel('Intensidad de pixel')
plt.ylabel('Frecuencia')
```

Algoritmo de Clahe

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread('Shar2.jpg', cv2.IMREAD_GRAYSCALE)
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(9,9))
new_img=clahe.apply(img_gray)

histogram,_ = np.histogram(img_gray.flatten(),256,[0,256])
histogram2,__ = np.histogram(new_img.flatten(),256,[0,256])

plt.figure(figsize=(12, 9))

plt.subplot(221)
plt.imshow(img_gray,cmap="gray")
plt.title("Imagen Original")

plt.subplot(222)
plt.imshow(new_img,cmap="gray")
plt.title("Imagen Equalizada")

plt.subplot(223)
plt.plot(histogram, color = "purple")
plt.title('HISTOGRAMA')
plt.xlabel('Intensidad de pixel')
plt.ylabel('Frecuencia')

plt.subplot(224)
plt.plot(histogram2, color = "purple")
plt.title('HISTOGRAMA EQUALIZADO')
plt.xlabel('Intensidad de pixel')
plt.ylabel('Frecuencia')
```
