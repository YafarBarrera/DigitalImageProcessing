Ejercicio 1 - Recorte de Zona de Inter√©s de una Imagen

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Imagen1.jpg') #Cargo la Imagen

img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Convierto la imagen de BGR a RGB

img_zone=img_rgb[610:800,660:840,:] #Recorto la Imagen
plt.imsave('Capilla.jpg',img_zone) #Guardo la Imagen
celula_cancer_img=plt.imread('Capilla.jpg') #Cargo la Imagen Guardada


plt.figure(figsize=(12,6)) #Creo la Figura para mostrar los resultados
#Muestro la imagen original
plt.subplot(121)
plt.imshow(img_rgb)
plt.title("Imagen Original")
#Muestro la zona de inter√©s
plt.subplot(122)
plt.imshow(img_zone)
plt.title("Zona de Inter√©s")

plt.show()
```

Ejercicio 5 - Aplicaci√≥n de Degradado

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import copy


def degrader(img):
    hight = len(img)
    width = len(img[0])
    new_img=copy.deepcopy(img)
    cen=hight/256
    cen2=cen
    flag=0
    for i in range(0,hight):
        for j in range(0,width):
            if(i<cen2):
                if(img[i][j]-flag<0):
                    new_img[i][j]=0
                else:
                    new_img[i][j]=img[i][j]-flag
                if(i==cen2-1):
                    cen2=cen2+cen
                    flag=flag+1
    return new_img

img = cv2.imread('Imagen1.jpg',cv2.IMREAD_GRAYSCALE) #Cargo la Imagen

img_degrad=degrader(img)

plt.figure(figsize=(12,6)) #Creo la Figura para mostrar los resultados
#Muestro la imagen original
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title("Imagen Original")
#Muestro la zona de inter√©s
plt.subplot(122)
plt.imshow(img_degrad, cmap='gray')
plt.title("Imagen Degradada")

plt.show()
```
